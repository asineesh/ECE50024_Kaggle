```python
import csv
import random
from torch.utils.data import random_split
```

```python
#Saving the label for each image name
lines = []
with open('train.csv', mode ='r') as file:
    csvFile = csv.reader(file)
    for line in csvFile:
        if line[0].isnumeric():
            lines.append(line)
```

```python
len(lines)
```

```python
random.shuffle(lines)
lines
```

```python
train_set = lines[0:int(0.8*len(lines))]
test_set = lines[int(0.8*len(lines)):]
```

```python
len(train_set)
```

```python
len(test_set)
```

```python
with open('main_train.csv', mode ='w',newline='') as file:
    writer = csv.writer(file)
    for line in train_set:
        writer.writerow(line)    

file.close()
```

```python
with open('main_test.csv', mode ='w',newline='') as file:
    writer = csv.writer(file)
    for line in test_set:
        writer.writerow(line)    

file.close()
```
